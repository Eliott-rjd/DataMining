Collect de données

```python
#Enregistrement des données en JSON
data=dict()

with open('data.json', 'a') as mon_fichier:
  for i in range(500):
    imgfile = Image.open("images/"+str(i)+".png")
    if(i<250):
      classe = "chat"
      id = i
    else:
      classe = "chien"
      id = i - 250
    
    data[i] = {
              "class": classe,
              "id" : id,
              "size": imgfile.size,
              "formar": imgfile.format,
              "mode": imgfile.mode,
          }
  json.dump(data, mon_fichier, sort_keys=True, indent=4)
  mon_fichier.close()
```
