```python
import os,sys
import json
from PIL import Image
import matplotlib.pyplot as plot
import numpy
import math
from sklearn.cluster import MiniBatchKMeans
from sklearn.cluster import KMeans
from webcolors import CSS3_HEX_TO_NAMES, hex_to_rgb
from scipy.spatial import KDTree
```

Collect de données

```python
#Enregistrement des données en JSON
data=dict()

with open('data2.json', 'a') as mon_fichier:
  for i in range(500):
    imgfile = Image.open("images/"+str(i)+".png")
    if(i<250):
      classe = "chat"
      id = i
    else:
      classe = "chien"
      id = i - 250
    
    data[i] = {
              "class": classe,
              "id" : id,
              "size": imgfile.size,
              "formar": imgfile.format,
              "mode": imgfile.mode,
          }
  json.dump(data, mon_fichier, sort_keys=True, indent=4)
mon_fichier.close()
```

```python
#Permet de retourner le nom de la couleur en fonction des valeurs RGB
def convert_rgb_to_names(rgb_tuple):
    
    # a dictionary of all the hex and their respective names in css3
    css3_db = CSS3_HEX_TO_NAMES
    names = []
    rgb_values = []    
    for color_hex, color_name in css3_db.items():
        names.append(color_name)
        rgb_values.append(hex_to_rgb(color_hex))
    
    kdt_db = KDTree(rgb_values)    
    distance, index = kdt_db.query(rgb_tuple)
    return names[index]


#COnvertion hexadcimal à rgb pour pouvoir utilisé la fonction au dessus
def hex_to_rgb(value):
    value = value.lstrip('#')
    lv = len(value)
    return tuple(int(value[k:k + lv // 3], 16) for k in range(0, lv, lv // 3))
```

```python
with open('data2.json', 'r') as mon_fichier:
  read = json.load(mon_fichier)
mon_fichier.close()

with open('data2.json', 'w') as mon_fichier:
  cluster_count = 3

  for i in range(500):
    imgfile = Image.open("images/"+str(i)+".png")
    numarray = numpy.array(imgfile.getdata(), numpy.uint8)
    clusters = MiniBatchKMeans(n_clusters = cluster_count)
    clusters.fit(numarray)

    nom_couleur = []
    for j in range(cluster_count):
        couleur = ('#%02x%02x%02x' % (
        math.ceil(clusters.cluster_centers_[j][0]), 
            math.ceil(clusters.cluster_centers_[j][1]),
        math.ceil(clusters.cluster_centers_[j][2])))

        couleur_rgb = hex_to_rgb(couleur)
        nom_couleur.append(convert_rgb_to_names(couleur_rgb))
        
    read[str(i)]["couleur"] = nom_couleur
    print(i, nom_couleur)
  json.dump(read, mon_fichier, sort_keys=True, indent=4)

mon_fichier.close()
```
